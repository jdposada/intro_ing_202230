```python
import pandas as pd
enlace = "https://www.datos.gov.co/api/views/rnvb-vnyh/rows.csv?accessType=DOWNLOAD"
icfes = pd.read_csv(enlace)
```

```python
icfes.shape
```

```python
cond = (icfes['ESTU_DEPTO_RESIDE'].isin(['MAGDALENA', 'ATLANTICO', 'CUNDINAMARCA']))
icfes_sub = icfes.loc[cond, ['ESTU_DEPTO_RESIDE', 'PUNT_INGLES', 'PUNT_GLOBAL']]
```

```python
icfes_sub.shape
```

```python
icfes_sub.head()
```

Como obtengo el promedio del puntaje por departamento en ingl√©s y el total

```python
icfes_sub.groupby(by="ESTU_DEPTO_RESIDE").mean()
```

```python
icfes_sub.groupby(by="ESTU_DEPTO_RESIDE").min()
```

```python
icfes_sub.groupby(by="ESTU_DEPTO_RESIDE").max()
```
